```python
import pandas as pd

# Define the file path for the new CSV file
file_path = r'C:\Users\alee\Desktop\YouGov - all adults transposed.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Melt the DataFrame to have three columns: date, issue, and percentage
df_melted = df.melt(id_vars=['Date'], var_name='issue', value_name='percentage')

# Print the first few rows of the melted DataFrame
display(df_melted)
```

```python
# Export the melted DataFrame as a CSV file

df_melted.to_csv(r'C:\Users\alee\Desktop\melted_dataYouGovAll.csv', index=False)
```

```python
# Define the file path for the new CSV file
file_path = r'C:\Users\alee\Desktop\YouGov - Male Transposed.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Melt the DataFrame to have three columns: date, issue, and percentage
df_melted = df.melt(id_vars=['Date'], var_name='issue', value_name='percentage')

# Remove the percentage sign from the 'percentage' column
df_melted['percentage'] = df_melted['percentage'].str.rstrip('%').astype(float)

# Print the first few rows of the melted DataFrame
display(df_melted)
```
